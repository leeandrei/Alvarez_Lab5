```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline 
```

<h5>Task 1<i>

$$A = 4x + 2y + z$$

```python
A = np.array([4,2,1])
A

x = np.arange(0,100)
plt.xlim(0,10)
plt.ylim(0,10)
plt.plot(x*A[0], x*A[1], color = 'red')
plt.show()
```

<h5>Task 2<i>

$$ B = \left\{\begin{array}\\ 6x+3y-2z\\ 4x-2y+3z\\ 5x-3y+8z \end{array}\right. $$

```python
B = np.array([[6,3,-2],[4,-2,3],[5,-3,8]])

X = plt.figure()
X1 = X.gca(projection = '3d')
X1.set_xlim([0,10])
X1.set_ylim([0,10])
X1.set_zlim([0,10])

origin = (0,0,0)
X1.quiver(origin, origin, origin, B[:,0], B[:,1], B[:,2], arrow_length_ratio = 0.05, colors = ['blue','red','green'])
plt.show()
```
